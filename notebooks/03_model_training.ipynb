```python
import pandas as pd

# Load your feature CSV
df = pd.read_csv("Data/features_extracted.csv")
print(f"✅ Loaded dataset: {df.shape[0]} samples, {df.shape[1]} features")

# Inspect a few rows
display(df.head())

# Check missing values
print(df.isnull().sum().sum(), "missing values in total")

# Check label distribution
print(df['label'].value_counts())
```

```python
print("Shape:", df.shape)
```

```python
import numpy as np

# Select only numeric columns for these operations
numeric_df = df.select_dtypes(include=[np.number])

# Check how many missing or infinite values exist
missing_before = df.isnull().sum().sum()
infinite_before = np.isinf(numeric_df.values).sum()

print(f"Before cleaning: {missing_before} missing, {infinite_before} infinite values.")

# Replace infinities with NaN for uniform handling
df.replace([np.inf, -np.inf], np.nan, inplace=True)

# Fill NaN with the mean value of each numeric column
df.fillna(df.mean(numeric_only=True), inplace=True)

# Verify cleanup
numeric_df = df.select_dtypes(include=[np.number])  # reselect after cleanup
missing_after = df.isnull().sum().sum()
infinite_after = np.isinf(numeric_df.values).sum()

print(f"After cleaning: {missing_after} missing, {infinite_after} infinite values.")
print("✅ All missing/infinite values handled successfully.")
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize the LabelEncoder
label_encoder = LabelEncoder()

# Fit encoder and transform the 'label' column
df['label_encoded'] = label_encoder.fit_transform(df['label'])

# Display a few encoded labels
print(df[['label', 'label_encoded']].head(10))

# Save mapping for later reference
label_map = dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))

print("\n🎵 Label Mapping (Genre → Encoded):")
for genre, code in label_map.items():
    print(f"{genre:10s} → {code}")
```

```python
print("Unique encoded labels:", sorted(df['label_encoded'].unique()))
print("Total genres:", len(df['label_encoded'].unique()))
```

```python
from sklearn.preprocessing import StandardScaler
import joblib
import numpy as np

# Separate features (X) and labels (y)
# Drop non-feature columns ('label' and 'label_encoded')
X = df.drop(columns=['label', 'label_encoded'])
y = df['label_encoded']

# Initialize the scaler
scaler = StandardScaler()

# Fit the scaler on the features and transform
X_scaled = scaler.fit_transform(X)

# Convert back to a DataFrame for easy inspection
X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)

print(f"✅ Features scaled successfully.")
print(f"Mean of all features (approx): {np.mean(X_scaled):.3f}")
print(f"Standard deviation (approx): {np.std(X_scaled):.3f}")

# Save the fitted scaler for later reuse
joblib.dump(scaler, "models/feature_scaler.pkl")
print("📦 Scaler saved at models/feature_scaler.pkl")
```
